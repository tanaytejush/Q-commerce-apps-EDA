```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set default plot styles
sns.set(style="whitegrid")
plt.style.use("ggplot")

# Load the dataset
df = pd.read_csv("Ecommerce_Delivery_Analytics_New.csv")

# Preview the data
df.head()
```

```python
# Understand the Dataset Structure
```

```python
# Shape of the dataset
df.shape

# List of columns
df.columns

# Data types and non-null counts
df.info()

# Check for missing values
df.isnull().sum()

# Basic descriptive statistics
df.describe(include='all')
```

```python
# Univariate Analysis
```

```python
# Delivery time 
```

```python
# Histogram of delivery time
plt.figure(figsize=(8, 5))
sns.histplot(df['Delivery Time (Minutes)'], bins=30, kde=True)
plt.title("Distribution of Delivery Time")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

```python
# order value
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Order Value (INR)")
plt.title("Order Value Distribution")
plt.show()
```

```python
# service rating 
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Service Rating", palette="Set2")
plt.title("Service Rating Count")
plt.show()
```

```python
# Categorical Value Distribution
```

```python
# Platform distribution
sns.countplot(data=df, x="Platform", palette="Set1")
plt.title("Orders per Platform")
plt.xticks(rotation=45)
plt.show()
```

```python
# Delivery Delay
sns.countplot(data=df, x="Delivery Delay", palette="pastel")
plt.title("Delivery Delay Status")
plt.show()

# Refunds
sns.countplot(data=df, x="Refund Requested", palette="muted")
plt.title("Refund Requested Status")
plt.show()
```

```python
# Bivariate Analysis
```

```python
#  Delivery Time vs Platform
plt.figure(figsize=(10, 6))
sns.boxplot(x="Platform", y="Delivery Time (Minutes)", data=df)
plt.title("Delivery Time by Platform")
plt.ylabel("Delivery Time (Minutes)")
plt.xlabel("Platform")
plt.show()
```

```python
#Service Rating vs Delivery Delay
plt.figure(figsize=(8, 5))
sns.boxplot(x="Delivery Delay", y="Service Rating", data=df)
plt.title("Impact of Delivery Delay on Service Rating")
plt.show()

```

```python
# Order Value vs Refund Requested
plt.figure(figsize=(8, 5))
sns.boxplot(x="Refund Requested", y="Order Value (INR)", data=df)
plt.title("Order Value by Refund Request")
plt.show()

```

```python
# Product Category vs Delivery Time
plt.figure(figsize=(10, 6))
sns.boxplot(x="Product Category", y="Delivery Time (Minutes)", data=df)
plt.title("Delivery Time by Product Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Platform vs Service Rating
plt.figure(figsize=(8, 5))
sns.boxplot(x="Platform", y="Service Rating", data=df)
plt.title("Service Rating by Platform")
plt.show()
```

```python
# Delivery Delay vs Refund Requested
pd.crosstab(df['Delivery Delay'], df['Refund Requested'], normalize='index') * 100
```

```python
# GroupBy Analysis
# Average delivery time per platform
df.groupby("Platform")["Delivery Time (Minutes)"].mean()

# Refund rate by product category
df.groupby("Product Category")["Refund Requested"].value_counts(normalize=True).unstack()
```
